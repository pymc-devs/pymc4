```python
%load_ext autoreload
%autoreload 2
import pymc4 as pm
import tensorflow as tf
```

```python
Y = tf.random.normal((400,))
```

```python
@pm.model
def model():
    m = yield pm.Normal("m", 0, 1)
    p = yield pm.Normal("p", m, 1, observed=2.)
    return p
```

```python
final_res = pm.sample_smc(model())
```

```python
import matplotlib.pyplot as plt
plt.plot(final_res[0])
```

```python
trace = pm.sample(model(), progressbar=True)
```

```python
import arviz as az
az.plot_trace(trace);
```
