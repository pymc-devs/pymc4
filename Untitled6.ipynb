```python
%load_ext autoreload
%autoreload 2
import arviz as az

import pymc4 as pm
import tensorflow as tf
import numpy as np

hyper_mean = 0
hyper_scale=10
sigma = 0.2

dtype = tf.float32

@pm.model()
def model():
    x = yield pm.NormalMixture(
        "x",
        w = tf.constant([[0.5, 0.5],[0.5, 0.5]]),
        loc=tf.constant([[0., 10.],[0., 10.]]),
        scale=tf.constant([[0.5, 0.5],[0.5, 0.5]])
    )
```

```python
trace1 = pm.sample(model(), num_samples=1000, progressbar=True, step_size=1.)
```

```python
az.plot_trace(trace1)
```

```python
samples = pm.NormalMixture.dist(
    w = np.array([0.2, 0.8]),
    loc=np.array([0., 10.]),
    scale=np.array([0.5, 0.5])
).sample(1000)
```

```python
import seaborn as sns
sns.distplot(samples)
```

```python
import numpy as np
import pymc3 as pm3
with pm3.Model() as model:
    x_obs = pm3.NormalMixture('x_obs', w=np.array([0.5, 0.5]), mu=np.array([0., 10.]), tau=np.array([0.5, 0.5]))
    trace = pm3.sample(1000, n_init=10000, tune=1000, chains=10)
```
