```python
%load_ext autoreload
%autoreload 2

import pymc4 as pm
import tensorflow as tf
import numpy as np
```

```python
x1 = tf.random.normal((100,))*1+0
x2 = tf.random.normal((300,))*1+4
x = tf.concat((x1, x2), 0)
```

```python
@pm.model()
def model():
    mean2 = yield pm.Normal("var", 0, 2)
    mk = yield pm.Mixture("x", tf.constant([0.27, 0.75]), [pm.Normal.dist(0,1), pm.Normal.dist(mean2,1)], observed=x)
    return mk
```

```python
trace = pm.sample(model(), progressbar=True)
```

```python
import arviz as az
az.plot_trace(trace)
```
