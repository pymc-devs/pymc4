```python
import pymc4 as pm
from pymc4.mcmc.samplers import NUTS, RandomWalkM
import arviz as az
```

```python
@pm.model
def model():
    bernoulli = yield pm.Bernoulli("ber", 0.2)
```

```python
trace = pm.sample(model(), sampler_type="compound", sampler_methods=[("ber", RandomWalkM)])
```

```python
az.plot_trace(trace)
```
