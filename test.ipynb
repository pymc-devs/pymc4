```python
import pymc4 as pm
import tensorflow_probability as tfp
tfd = tfp.distributions
tfb = tfp.distributions.bijectors
ed = tfp.edward2
```

```python
from pymc4.random_variable import RandomVariable as RV
import tensorflow as tf
import seaborn as sns
import matplotlib.pyplot as plt
import xarray as xr
```

## Arithematic Operations
### Addition

```python
with pm.Model() as model:
    x = RV(tfd.Normal(0., 1.), name="x", sample_shape=10000)
    y = RV(tfd.Normal(1., 1.), name="y", sample_shape=10000)
    z = x + y
with tf.Session() as sess:
    sns.distplot(sess.run(z))
    plt.plot()
    sns.distplot(sess.run(x))
    plt.plot()
    sns.distplot(sess.run(y))
    plt.plot()
```

### Multiplication

```python
with pm.Model() as model:
    x = RV(tfd.Normal(0., 1.), name="x", sample_shape=10000)
    y = RV(tfd.Normal(1., 1.), name="y", sample_shape=10000)
    z = x * y
with tf.Session() as sess:
    sns.distplot(sess.run(z))
    plt.plot()
    sns.distplot(sess.run(x))
    plt.plot()
    sns.distplot(sess.run(y))
    plt.plot()
```

### Substraction

```python
with pm.Model() as model:
    x = RV(tfd.Normal(0., 1.), name="x", sample_shape=10000)
    y = RV(tfd.Normal(1., 1.), name="y", sample_shape=10000)
    z = x - y
with tf.Session() as sess:
    sns.distplot(sess.run(z))
    plt.plot()
    sns.distplot(sess.run(x))
    plt.plot()
    sns.distplot(sess.run(y))
    plt.plot()
```
